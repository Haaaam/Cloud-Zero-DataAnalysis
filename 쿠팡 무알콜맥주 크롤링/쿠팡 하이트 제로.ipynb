## 쿠팡 하이트제로 리뷰가져오기

```python
html=driver.page_source
soup=bs(html,'html.parser')

driver.execute_script("window.scrollTo(2, document.body.scrollHeight);")
pg=200
cnt=0
data=[]


#pn=int(soup.find('button',class_="sdp-review__article__page__num sdp-review__article__page__num--active js_reviewArticlePageBtn")["data-page"])
for i in range(1,pg):
    html=driver.page_source
    soup=bs(html,'html.parser')
    
    
    #전체 리뷰 갯수
    total=int(soup.find("div",class_='sdp-review__average__total-star__info-count').text.replace(',',''))
    
    
    #리뷰 가져오기
    temp=soup.find('div', class_="sdp-review")
    review_lst=temp.find_all('div',class_="sdp-review__article__list__review__content js_reviewArticleContent")
    
    time.sleep(3)
    
    for j in range(5):
        cnt+=1
        #review_id=j.find('span').text
        try:
            review=review_lst[j].text
        except:
            break
        
       
        print(review)
        data.append(review)
        
        if cnt==total:
            break
    if cnt==total:
        break

driver.close()
```

```python
df=pd.DataFrame(data)
#print(data)
df.to_csv('./data/하이트제로 크롤링.csv')
```

```python
from konlpy.tag import Okt
import os
import nltk
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
review_txt=""
for txt in data:
    if txt!='':
        review_txt+=txt+"\n"
okt=Okt()
noun_txt=okt.nouns(review_txt)
noun_ko=nltk.Text(noun_txt)
print(noun_ko)
```

```python
noun=okt.nouns(review_txt)
noun
```

```python
cnt=Counter(noun)
cnt_rank=cnt.most_common()
cnt_rank
```

```python
stop_words=['맥주','때','술','캔','진짜','기분','주문','마트','거','것','좀','이',
           '느낌','취하','처음','분','전','제','정도','한잔','요','그냥','또','대신',
           '사람','아주','지금','걸','중','계속','수','때문','개','달','나','후','별로',
           '해','못','하나','냉장고','안','듯','쿠팡','저','더','가격','배송','정말','구매']
word_txt=[txt for txt in noun_txt if txt not in stop_words]
word_txt
cnt=Counter(word_txt)
cnt=cnt.most_common()
cnt
```

```python
rank_cnt=[]

for (i,j) in cnt:
    if j>80:
        rank_cnt.append((i,j))
                        
rank_dic=dict(rank_cnt)

rank_cnt2=[]

for (i,j) in cnt:
    if j>30:
        rank_cnt2.append((i,j))
rank_dic2=dict(rank_cnt2)
```

```python
wc = WordCloud(font_path = "c:/windows/fonts/malgun.ttf"
              , background_color = "black", width = 600, height = 600
              , max_words = 100, max_font_size = 300
              ).generate_from_frequencies(rank_dic2)

plt.figure(figsize = (12,12))
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
df = pd.DataFrame(rank_cnt, columns = ['단어', '빈도수'])
df
```
